# Predicting LLM Activations with QA-Emb

This notebook extends the QA-Emb paper (Benara et al., 2024) to predict **LLM internal activations** instead of fMRI voxels. Uses the same fMRI story corpus (OpenNeuro ds003020) for direct comparison with brain results.

**Pipeline:**
1. Prepare text n-grams from the full fMRI story corpus (same stories used in the paper)
2. Extract Llama 3-8B hidden states at layers 0 (first), 16 (middle), 31 (last)
3. Compute QA-Emb: answer 674 yes/no questions per n-gram
4. PCA on activations → top 100 components (matching fMRI paper)
5. Ridge regression: QA-Emb → PCA components (story-based train/test split)
6. Evaluate: can interpretable yes/no questions predict LLM representations?

## 6. Extract Activations

For each n-gram, we extract the **last token's hidden state** at layers 0, 16, and 31. Batch size is 8 to avoid GPU OOM (storing all 33 layer hidden states is memory-intensive). Checkpoints to `CACHE_DIR` every 512 samples.

```python
LAYER_INDICES = [0, num_layers // 2, num_layers - 1]  # [0, 16, 31]
BATCH_SIZE_ACT = 8  # reduced from 32 to avoid OOM with output_hidden_states
CHECKPOINT_EVERY = 512

cache_file = os.path.join(CACHE_DIR, 'activations_full.npz')
partial_file = os.path.join(CACHE_DIR, 'activations_full_partial.npz')

if os.path.exists(cache_file):
    print('Loading cached activations...')
    cached = np.load(cache_file)
    activations = {int(k.split('_')[1]): cached[k] for k in cached.files}
else:
    n_total = len(ngrams)
    torch.cuda.empty_cache()

    # Check for partial progress
    if os.path.exists(partial_file):
        cached = np.load(partial_file)
        start = int(cached['progress'])
        acts = {int(k.split('_')[1]): cached[k] for k in cached.files if k != 'progress'}
        for l in LAYER_INDICES:
            full = np.zeros((n_total, hidden_size), dtype=np.float32)
            full[:start] = acts[l]
            acts[l] = full
        print(f'Resuming activation extraction from sample {start}/{n_total}')
    else:
        start = 0
        acts = {l: np.zeros((n_total, hidden_size), dtype=np.float32) for l in LAYER_INDICES}

    for i in tqdm(range(start, n_total, BATCH_SIZE_ACT), desc='Extracting activations'):
        batch = ngrams[i:i + BATCH_SIZE_ACT]
        inputs = tokenizer(
            batch, return_tensors='pt', padding=True,
            truncation=True, max_length=128,
        ).to(device)

        with torch.no_grad():
            outputs = model(**inputs, output_hidden_states=True)

        hidden_states = outputs.hidden_states
        seq_lengths = inputs['attention_mask'].sum(dim=1) - 1

        for l in LAYER_INDICES:
            h = hidden_states[l + 1][torch.arange(len(batch)), seq_lengths]
            acts[l][i:i + len(batch)] = h.cpu().float().numpy()

        # Free GPU memory
        del outputs, hidden_states, inputs
        torch.cuda.empty_cache()

        # Periodic checkpoint
        done = i + len(batch)
        if done % CHECKPOINT_EVERY < BATCH_SIZE_ACT or done == n_total:
            save_dict = {f'layer_{l}': acts[l][:done] for l in LAYER_INDICES}
            save_dict['progress'] = np.array(done)
            np.savez(partial_file, **save_dict)

    activations = acts
    np.savez(cache_file, **{f'layer_{l}': v for l, v in activations.items()})
    if os.path.exists(partial_file):
        os.remove(partial_file)
    print('Activations saved to final cache.')

for l in LAYER_INDICES:
    print(f'Layer {l}: shape {activations[l].shape}')
```

## 7. Load Precomputed QA-Emb (Llama-3-8B)

Single-model setup: load the Llama-3-8B-Instruct answers from the precomputed cache, concatenate train + test, and cast bool → float32. No ensemble, no API calls.

```python
qa_train = np.load(f'{LLAMA_FOLDER}/v3_boostexamples_answers_train_numpy.npz')['arr_0']
qa_test  = np.load(f'{LLAMA_FOLDER}/v3_boostexamples_answers_test_numpy.npz')['arr_0']

assert qa_train.shape == (184564, 674), qa_train.shape
assert qa_test.shape  == (5951, 674),   qa_test.shape

qa_emb = np.concatenate([qa_train, qa_test], axis=0).astype(np.float32)
del qa_train, qa_test

N_QUESTIONS = qa_emb.shape[1]
print(f'QA-Emb shape: {qa_emb.shape}')
print(f'Mean yes-rate: {qa_emb.mean():.3f}')
print(f'Per-question yes-rate range: [{qa_emb.mean(axis=0).min():.3f}, {qa_emb.mean(axis=0).max():.3f}]')
```

## 8. PCA on Activations

For each layer, we run PCA on the activation matrix and extract the top 100 components (matching the fMRI paper, now justified with ~27k samples). PCA is fit on training data only to avoid leakage.

```python
N_PCA_COMPONENTS = 100

# Story-based train/test split (matching the fMRI paper)
n_train = len(train_ngrams)
n_test = len(test_ngrams)
train_idx = np.arange(n_train)
test_idx = np.arange(n_train, n_train + n_test)
print(f'Train: {n_train}, Test: {n_test} (story-based split)')

pca_models = {}
pca_activations = {}

pca_cache_file = os.path.join(CACHE_DIR, 'pca_components.npz')
pca_var_cache_file = os.path.join(CACHE_DIR, 'pca_explained_var.npz')
pca_cache_has_var = os.path.exists(pca_var_cache_file)

if os.path.exists(pca_cache_file):
    print(f'Loading cached PCA fits from {pca_cache_file}')
    pca_cache = np.load(pca_cache_file)
    pca_var_cache = np.load(pca_var_cache_file) if pca_cache_has_var else None

    fig, axes = plt.subplots(1, len(LAYER_INDICES), figsize=(15, 4))
    evr_to_save = {}
    for ax_idx, layer_idx in enumerate(LAYER_INDICES):
        components = pca_cache[f'components_layer_{layer_idx}'].astype(np.float32)
        mean = pca_cache[f'mean_layer_{layer_idx}'].astype(np.float32)
        class _PCALike: pass
        pca = _PCALike()
        pca.components_ = components
        pca.mean_ = mean
        pca.inverse_transform = lambda Z, c=components, m=mean: Z @ c + m

        # Compute PCA projection FIRST so EVR has access to Z.
        Z = (activations[layer_idx] - mean) @ components.T   # (n, 100)
        pca_models[layer_idx] = pca
        pca_activations[layer_idx] = Z

        if pca_var_cache is not None:
            pca.explained_variance_ratio_ = pca_var_cache[f'evr_layer_{layer_idx}']
        else:
            # Inline EVR from train-set Z and centered train activations.
            Z_train = Z[train_idx]
            explained_var = Z_train.var(axis=0, ddof=1)
            train_centered_var = (activations[layer_idx][train_idx]
                                    - mean).var(axis=0, ddof=1).sum()
            pca.explained_variance_ratio_ = (
                explained_var / train_centered_var
            ).astype(np.float32)
            evr_to_save[f'evr_layer_{layer_idx}'] = pca.explained_variance_ratio_

        cumvar = np.cumsum(pca.explained_variance_ratio_)
        axes[ax_idx].plot(range(1, N_PCA_COMPONENTS + 1), cumvar,
                          'b-o', markersize=2)
        axes[ax_idx].axhline(y=0.9, color='r', linestyle='--', alpha=0.5,
                              label='90% variance')
        axes[ax_idx].set_ylim(0, 1)
        axes[ax_idx].legend(loc='lower right', fontsize=9)
        axes[ax_idx].set_ylabel('Cumulative variance explained')
        axes[ax_idx].set_xlabel('# PCA components')
        axes[ax_idx].set_title(f'Layer {layer_idx}')
        print(f'Layer {layer_idx}: PCA shape {Z.shape} (loaded from cache)')

    if evr_to_save:
        np.savez(pca_var_cache_file, **evr_to_save)
        print(f'Saved {pca_var_cache_file} for future runs.')

    plt.suptitle('PCA: cumulative variance explained per layer', fontsize=14)
    plt.tight_layout()
    plt.show()
else:
    fig, axes = plt.subplots(1, len(LAYER_INDICES), figsize=(15, 4))
    components_save = {}; mean_save = {}; evr_save = {}
    for ax_idx, layer_idx in enumerate(LAYER_INDICES):
        pca = PCA(n_components=N_PCA_COMPONENTS)
        pca.fit(activations[layer_idx][train_idx])
        pca_act = pca.transform(activations[layer_idx])
        pca_models[layer_idx] = pca
        pca_activations[layer_idx] = pca_act

        cumvar = np.cumsum(pca.explained_variance_ratio_)
        axes[ax_idx].plot(range(1, N_PCA_COMPONENTS + 1), cumvar, 'b-o', markersize=2)
        axes[ax_idx].axhline(y=0.9, color='r', linestyle='--', alpha=0.5,
                              label='90% variance')
        axes[ax_idx].set_xlabel('# PCA components')
        axes[ax_idx].set_ylabel('Cumulative variance explained')
        axes[ax_idx].set_title(f'Layer {layer_idx}')
        axes[ax_idx].legend(loc='lower right', fontsize=9)
        axes[ax_idx].set_ylim(0, 1)

        components_save[f'components_layer_{layer_idx}'] = pca.components_.astype(np.float32)
        mean_save[f'mean_layer_{layer_idx}'] = pca.mean_.astype(np.float32)
        evr_save[f'evr_layer_{layer_idx}'] = pca.explained_variance_ratio_.astype(np.float32)
        print(f'Layer {layer_idx}: PCA shape {pca_act.shape}, '
              f'cum. variance ({N_PCA_COMPONENTS} PCs) = {cumvar[-1]:.3f}')

    np.savez(pca_cache_file, **components_save, **mean_save)
    np.savez(pca_var_cache_file, **evr_save)
    print(f'Saved PCA cache to {pca_cache_file} and {pca_var_cache_file}')

    plt.suptitle('PCA: cumulative variance explained per layer', fontsize=14)
    plt.tight_layout()
    plt.show()
```

## 9. Ridge Regression: QA-Emb → PCA Components

For each layer, we fit a ridge regression from the 674-dim QA-Emb to the 100-dim PCA activation space. We use `RidgeCV` with built-in cross-validation for alpha selection. Train/test split is story-based (matching the fMRI paper).

```python
ALPHAS = np.logspace(-1, 4, 20)  # Ridge regularization candidates

X_train = qa_emb[train_idx]
X_test = qa_emb[test_idx]

ridge_cache = os.path.join(CACHE_DIR, 'ridge_results.npz')

if os.path.exists(ridge_cache):
    print(f'Loading cached Ridge results from {ridge_cache}')
    cached = np.load(ridge_cache, allow_pickle=True)
    results = cached['results'].item()
    for layer_idx in LAYER_INDICES:
        r = results[layer_idx]
        print(f'Layer {layer_idx}: mean r = {r["mean_corr"]:.4f}, '
              f'best alpha = {r["best_alpha"]:.2f}, '
              f'top-5 corrs = {np.sort(r["correlations"])[::-1][:5].round(3)}')
else:
    results = {}
    for layer_idx in LAYER_INDICES:
        Y_train = pca_activations[layer_idx][train_idx]
        Y_test = pca_activations[layer_idx][test_idx]

        ridge = RidgeCV(alphas=ALPHAS, scoring='neg_mean_squared_error')
        ridge.fit(X_train, Y_train)

        Y_pred = ridge.predict(X_test)

        correlations = []
        for comp in range(N_PCA_COMPONENTS):
            if np.std(Y_test[:, comp]) > 1e-10 and np.std(Y_pred[:, comp]) > 1e-10:
                r, _ = pearsonr(Y_test[:, comp], Y_pred[:, comp])
                correlations.append(r)
            else:
                correlations.append(0.0)
        correlations = np.array(correlations)

        results[layer_idx] = {
            'coef': ridge.coef_,
            'correlations': correlations,
            'mean_corr': float(np.mean(correlations)),
            'best_alpha': float(ridge.alpha_),
            'Y_pred': Y_pred,
            'Y_test': Y_test,
        }

        print(f'Layer {layer_idx}: mean r = {np.mean(correlations):.4f}, '
              f'best alpha = {ridge.alpha_:.2f}, '
              f'top-5 corrs = {np.sort(correlations)[::-1][:5].round(3)}')

    np.savez(ridge_cache, results=results)
    print(f'Saved Ridge results to {ridge_cache}')
```

## 11. Feature Selection: How Few Questions Suffice?

Mirroring the paper's finding that 29 questions beat the 985-dim Eng1000 baseline, we test how QA-Emb performance degrades as we use fewer questions (selected by Elastic Net).

```python
from sklearn.linear_model import MultiTaskElasticNet

# Paper spec (Benara et al. 2024, §4.1 + footnote 2): MultiTaskElasticNet (no CV)
# over 20 log-spaced alphas in [1e-3, 1] with l1_ratio=0.5, then refit Ridge on
# the selected questions. The 29-question result is one specific alpha point.


def _per_col_pearson(X, Y):
    Xc = X - X.mean(axis=0, keepdims=True)
    Yc = Y - Y.mean(axis=0, keepdims=True)
    num = (Xc * Yc).sum(axis=0)
    den = np.sqrt((Xc * Xc).sum(axis=0) * (Yc * Yc).sum(axis=0))
    den[den < 1e-12] = 1.0
    return num / den


def analyze_layer(ANALYSIS_LAYER):
    cache_path = os.path.join(CACHE_DIR, f'analyze_layer_{ANALYSIS_LAYER}.npz')
    partial_path = os.path.join(CACHE_DIR, f'analyze_layer_{ANALYSIS_LAYER}_partial.npz')

    if os.path.exists(cache_path):
        print(f'Loading cached analyze_layer({ANALYSIS_LAYER}) from {cache_path}')
        cache = np.load(cache_path)
        alphas_path = cache['alphas']
        n_nonzero_arr = cache['n_nonzero']
        perf_arr_pca = cache['perf']        # PCA-space mean r per alpha
    else:
        Y_train_a = pca_activations[ANALYSIS_LAYER][train_idx]
        Y_test_a = pca_activations[ANALYSIS_LAYER][test_idx]
        alphas_path = np.logspace(0, -3, 20)
        n_alphas = len(alphas_path)

        if os.path.exists(partial_path):
            p = np.load(partial_path, allow_pickle=True)
            done_n = int(p['done'])
            n_nonzero_list = list(p['n_nonzero'][:done_n])
            perf_list = list(p['perf'][:done_n])
            saved_coef = p['enet_coef'] if 'enet_coef' in p.files else None
            saved_intercept = p['enet_intercept'] if 'enet_intercept' in p.files else None
            print(f'Resuming layer {ANALYSIS_LAYER} from alpha {done_n}/{n_alphas}')
        else:
            done_n = 0; n_nonzero_list = []; perf_list = []
            saved_coef = None; saved_intercept = None

        enet = MultiTaskElasticNet(l1_ratio=0.5, max_iter=2000, tol=1e-3, warm_start=True)
        if saved_coef is not None:
            enet.coef_ = saved_coef; enet.intercept_ = saved_intercept
            enet.n_iter_ = 0; enet.n_features_in_ = saved_coef.shape[1]
            enet.dual_gap_ = 0.0

        for i in tqdm(range(done_n, n_alphas), desc=f'Layer {ANALYSIS_LAYER} ENet path'):
            alpha = alphas_path[i]
            enet.alpha = float(alpha)
            enet.fit(X_train, Y_train_a)
            row_norms = np.linalg.norm(enet.coef_, axis=0)
            selected_mask = row_norms > 1e-10
            n_selected = int(selected_mask.sum())
            if n_selected == 0:
                n_nonzero_list.append(0); perf_list.append(0.0)
            else:
                X_tr_sub = X_train[:, selected_mask]; X_te_sub = X_test[:, selected_mask]
                ridge_sub = RidgeCV(alphas=ALPHAS); ridge_sub.fit(X_tr_sub, Y_train_a)
                Y_pred_sub = ridge_sub.predict(X_te_sub)
                corrs_sub = []
                for comp in range(N_PCA_COMPONENTS):
                    if np.std(Y_test_a[:, comp]) > 1e-10 and np.std(Y_pred_sub[:, comp]) > 1e-10:
                        r, _ = pearsonr(Y_test_a[:, comp], Y_pred_sub[:, comp])
                        corrs_sub.append(r)
                    else:
                        corrs_sub.append(0.0)
                n_nonzero_list.append(n_selected); perf_list.append(float(np.mean(corrs_sub)))
            done = i + 1
            np.savez(partial_path, done=np.array(done), alphas=alphas_path,
                     n_nonzero=np.array(n_nonzero_list + [0]*(n_alphas-done)),
                     perf=np.array(perf_list + [0.0]*(n_alphas-done)),
                     enet_coef=enet.coef_, enet_intercept=enet.intercept_)
        n_nonzero_arr = np.array(n_nonzero_list); perf_arr_pca = np.array(perf_list)
        np.savez(cache_path, alphas=alphas_path,
                 n_nonzero=n_nonzero_arr, perf=perf_arr_pca)
        if os.path.exists(partial_path):
            os.remove(partial_path)
        print(f'Saved analyze_layer({ANALYSIS_LAYER}) to {cache_path}')

    # === Full-activation correlation per ENet alpha point. ===
    # Interpolate from the precomputed top-K-by-Ridge-importance curve in
    # extra_metrics.npz. This is fast (no Ridge refit) and consistent with
    # the top-K sweep plot.
    extra = np.load(os.path.join(CACHE_DIR, "extra_metrics.npz"))
    K_vals = extra["K_values"]
    full_at_K = extra[f"topK_full_corr_layer_{ANALYSIS_LAYER}"]
    pca_at_K = extra[f"topK_pca_corr_layer_{ANALYSIS_LAYER}"]
    # Clip ENet n_nonzero values to the interpolation range
    nq_clipped = np.clip(n_nonzero_arr, K_vals.min(), K_vals.max())
    perf_arr_full = np.interp(nq_clipped, K_vals, full_at_K).astype(np.float32)
    # Overwrite PCA curve too so both curves use the same Ridge-importance
    # ranking (apples-to-apples comparison; matches the top-K plot).
    perf_arr_pca_interp = np.interp(nq_clipped, K_vals, pca_at_K).astype(np.float32)

    # === Plot: full activation correlation as primary, PCA as secondary ===
    mask = n_nonzero_arr >= 1
    nq_valid = n_nonzero_arr[mask]
    pca_valid = perf_arr_pca_interp[mask]
    full_valid = perf_arr_full[mask]
    n_dropped = int((~mask).sum())

    order = np.argsort(nq_valid)
    nq_sorted = nq_valid[order]
    pca_sorted = pca_valid[order]
    full_sorted = full_valid[order]

    fig, ax = plt.subplots(figsize=(9.5, 5))
    if len(nq_sorted) == 0:
        ax.text(0.5, 0.5, 'No alpha in the path selected >0 questions.',
                ha='center', va='center', transform=ax.transAxes)
    else:
        ax.plot(nq_sorted, full_sorted, 's-', color='#e15759', markersize=8,
                linewidth=2.2, label='Full activation r (4096 dims) — primary')
        ax.plot(nq_sorted, pca_sorted, 'o--', color='#4e79a7', markersize=6,
                linewidth=1.4, alpha=0.7, label='PCA-target r (100 dims) — secondary')

        if 29 >= nq_sorted.min() and 29 <= nq_sorted.max():
            ax.axvline(x=29, color='black', linestyle=':', linewidth=1.5, alpha=0.8,
                       label='29 questions (paper threshold)')
        elif 29 < nq_sorted.min():
            ax.axvline(x=29, color='black', linestyle=':', linewidth=1.2, alpha=0.5,
                       label=f'29 q (paper) — outside ENet range [{nq_sorted.min()}, {nq_sorted.max()}]')

        ax.set_xscale('log')
        ax.set_xlim(left=max(0.9, nq_sorted.min() * 0.85))

    ax.set_xlabel('# questions selected by ElasticNet (alpha-driven)')
    ax.set_ylabel('Mean test Pearson r')
    note = f' — {n_dropped} alpha(s) selected 0 questions, dropped' if n_dropped else ''
    ax.set_title(f'Layer {ANALYSIS_LAYER}: ElasticNet sparsity sweep{note}')
    ax.legend(loc='lower right')
    ax.grid(True, alpha=0.3, which='both')
    ax.set_ylim(bottom=0)
    plt.tight_layout()
    plt.show()

    print(f'\n  #q  |  full-act r  |  PCA-target r')
    print('-' * 42)
    for nq, p_full, p_pca in zip(nq_sorted, full_sorted, pca_sorted):
        print(f'  {int(nq):4d}  |   {p_full:.4f}   |    {p_pca:.4f}')
    if n_dropped:
        print(f'  ({n_dropped} alpha(s) selected 0 questions and are not shown.)')
```

## 11b. How few questions are needed? (Top-K sweep)

Rank the 674 questions by mean |Ridge weight| and refit using the top K. This lets us evaluate **exactly** K=29 (the paper threshold) and lets us compute the **full activation space** metric directly. The full-act curve (red, primary) reports mean per-dim Pearson r over the **4096-dim Llama activations**; the PCA curve (blue, secondary) is included for comparison with the standard QA-Emb evaluation.


```python
# Top-K (by Ridge importance) sweep, in PCA and full activation space.
# 29 questions is the paper's reported sparsity threshold.
extra = np.load(os.path.join(CACHE_DIR, 'extra_metrics.npz'))
K_values = extra['K_values']

fig, axes = plt.subplots(1, len(LAYER_INDICES), figsize=(18, 5.2), sharey=False)
K_list = K_values.tolist()

for ax_idx, layer_idx in enumerate(LAYER_INDICES):
    pca_curve = extra[f'topK_pca_corr_layer_{layer_idx}']
    full_curve = extra[f'topK_full_corr_layer_{layer_idx}']

    ax = axes[ax_idx]
    ax.plot(K_values, full_curve, 's-', color='#e15759',
            markersize=7, linewidth=2.2,
            label='Full activation r (4096 dims) — primary')
    ax.plot(K_values, pca_curve, 'o--', color='#4e79a7',
            markersize=6, linewidth=1.4, alpha=0.7,
            label='PCA-target r (100 dims) — secondary')
    ax.axvline(x=29, color='black', linestyle=':', linewidth=1.5, alpha=0.8,
               label='29 questions (paper)')

    if 29 in K_list:
        i29 = K_list.index(29)
        ax.plot(29, full_curve[i29], marker='*', markersize=20,
                color='#e15759', markeredgecolor='black', markeredgewidth=0.9,
                zorder=5, linestyle='None',
                label=f'@K=29: full r={full_curve[i29]:.3f}')
        ax.plot(29, pca_curve[i29], marker='*', markersize=16,
                color='#4e79a7', markeredgecolor='black', markeredgewidth=0.7,
                zorder=4, linestyle='None',
                label=f'@K=29: PCA r={pca_curve[i29]:.3f}')

    ax.set_xscale('log')
    ax.set_xlabel('# questions (top-K by mean |Ridge weight|)')
    if ax_idx == 0:
        ax.set_ylabel('Mean test Pearson r')
    ax.set_title(f'Layer {layer_idx}')
    ax.legend(loc='lower right', fontsize=8.5)
    ax.grid(True, alpha=0.3, which='both')
    ax.set_xlim(left=0.9, right=K_values.max() * 1.15)
    ax.set_ylim(bottom=0)

plt.suptitle('How few questions are needed? '
             '(top-K by Ridge importance; full activation r is the headline)',
             fontsize=14)
plt.tight_layout()
plt.show()

print('\n  K  | layer 0  full / PCA |  layer 16  full / PCA  |  layer 31  full / PCA')
print('-' * 84)
for i, K in enumerate(K_values):
    line = f'{int(K):4d}  '
    for layer_idx in LAYER_INDICES:
        p = extra[f'topK_pca_corr_layer_{layer_idx}'][i]
        f_ = extra[f'topK_full_corr_layer_{layer_idx}'][i]
        line += f'|  {f_:.4f} / {p:.4f}  '
    print(line)
```
